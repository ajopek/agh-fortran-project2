```python
import pandas as pan
import matplotlib.pyplot as plot
```

```python
# Load data
data = pan.read_table('result.data', sep=r"\s+", header=0)
```

```python
data['type'] = data.type.apply(lambda x: x.strip())
data2 = data.groupby(['n', 'type']).agg({'time': 'mean'})
data3 = data2.unstack('type')
data3.columns = data3.columns.get_level_values(1)
data3
```

```python
data3.plot(figsize=(10,15), loglog=True)
plot.savefig("graph.png")
plot.show()
```
