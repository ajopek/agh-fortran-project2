```python
import pandas as pan
import matplotlib.pyplot as plot
```

```python
# Load data
data_basic = pan.read_table('result_basic.data', sep=r"\s+", header=0)
data_dotproduct = pan.read_table('result_dotproduct.data', sep=r"\s+", header=0)
data_cached = pan.read_table('result_cached.data', sep=r"\s+", header=0)
data_both = pan.read_table('result_both.data', sep=r"\s+", header=0)
```

```python
data_basic.sort_values(by='n').plot(figsize=(15, 10), x='n', y='time')
plot.yscale('log')
plot.show()
```

```python
data_dotproduct.sort_values(by='n').plot(figsize=(15, 10), x='n', y='time')
plot.yscale('log')
plot.show()
```

```python
data_cached.sort_values(by='n').plot(figsize=(15, 10), x='n', y='time')
plot.yscale('log')
plot.show()
```

```python
data_both.sort_values(by='n').plot(figsize=(15, 10), x='n', y='time')
plot.yscale('log')
plot.show()
```
